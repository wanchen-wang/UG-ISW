```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        sorted_s = sorted(s)
        long_streak = 0
        for ss in sorted_s:
            hashmap = list()
            
            if ss not in hashmap:
                hashmap.append(ss)
                current_long = 1
                ss_next = chr(ord(ss) + 1)
                while ss_next not in hashmap:
                    current_long += 1
                    hashmap.append(ss_next)
                    ss_next = chr(ord(ss_next) + 1)
                    print(hashmap)
                    print(ss)
                    
                long_streak = max(long_streak, current_long)
            

        return long_streak
```
